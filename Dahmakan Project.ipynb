```python
from google.cloud import bigquery
from google.oauth2 import service_account
import pandas as pd
```

```python
credentials = service_account.Credentials.from_service_account_file('./projectCredentials.json')
```

```python
project_id = 'teak-perigee-259612'
```

```python
client = bigquery.Client(credentials=credentials,project=project_id)
```

```python
query_job = client.query("""SELECT * FROM (SELECT address.address, riderDistanceInMeters, (riderDepartedAt.DLnumberLong - riderArrivedAt.DLnumberLong)/60/1000 AS timeDifference, city, created_date.DLdate as createdDate, deliveredAt.DLnumberLong as deliveredAt, driver, estimatedArrivalTime.DLnumberLong as estimatedArrivalTime FROM `teak-perigee-259612.AllOrders.AllOrders` WHERE riderDistanceInMeters > 0) WHERE timeDifference > 0""")
```

```python
results = query_job.result()
```

```python
df = results.to_dataframe()
```
